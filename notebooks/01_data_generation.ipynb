```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 4)

T = 100                  # nombre de points
t = np.linspace(0, 1, T) # temps normalisé

def generate_sine(t, freq=2.0, amplitude=1.0, phase=0.0):
    return amplitude * np.sin(2 * np.pi * freq * t + phase)

signal = generate_sine(t, freq=3.0)
plt.plot(t, signal)
plt.title("Sinusoïde simple")
plt.show()


```

```python
def add_noise(signal, noise_std=0.1):
    noise = np.random.normal(0, noise_std, size=signal.shape)
    return signal + noise

noisy_signal = add_noise(signal, noise_std=0.2)

plt.plot(t, signal, label="signal propre")
plt.plot(t, noisy_signal, label="signal bruité", alpha=0.7)
plt.legend()
plt.title("Signal avec bruit")
plt.show()
```

```python
def generate_multi_sine(t, n_components=3):
    signal = np.zeros_like(t)
    for _ in range(n_components):
        freq = np.random.uniform(1, 5)
        amp = np.random.uniform(0.5, 1.5)
        phase = np.random.uniform(0, 2*np.pi)
        signal += amp * np.sin(2 * np.pi * freq * t + phase)
    return signal

signal = generate_multi_sine(t)
signal = add_noise(signal, noise_std=0.1)

plt.plot(t, signal)
plt.title("Somme de sinusoïdes bruitée")
plt.show()
```

```python
def generate_dataset(n_signals=1000, T=100):
    t = np.linspace(0, 1, T)
    signals = []

    for _ in range(n_signals):
        s = generate_multi_sine(t)
        s = add_noise(s, noise_std=0.1)
        signals.append(s)

    return np.array(signals)

signals = generate_dataset(n_signals=1000, T=100)
signals.shape
```

```python
for i in range(5):
    plt.plot(t, signals[i])
plt.title("Exemples de signaux générés")
plt.show()
```
